```python
# Importing the random and numpy modules
import random
import numpy as np

# Taking input from user for the total number of transactions
total_transactions = int(input("Enter the total number of transactions: "))

# Creating empty lists to store the transaction type and amount for each transaction
transaction_type = []
amount = []

# Looping over number of transactions
for i in range(total_transactions):
    # Take input for transaction type and amount for each transaction
    transaction, amount_new = input(f"Enter transaction {i+1}: ").split()
    transaction_type.append(transaction)
    amount_new = int(amount_new)
    if transaction == "d":
        amount.append(amount_new)
    else:
        amount.append(-amount_new)

# Setting the target sum to zero
target = 0


# Defining a function to generate a population of binary strings
def generate_population(total):
    population = []
    i = 0
    while i < 20:
        x = np.random.randint(2, size=total)
        population.append(x)
        i += 1
    return population


# Defining a function to calculate the fitness of each individual in the population
def calculate_fitness(population):
    scores = []
    i = 0
    while i < len(population):
        total = 0
        for idx, gene in enumerate(population[i]):
            if gene == 1:
                total += amount[idx]
        scores.append(abs(total))
        i += 1
    return scores





# Defining a function to select the most fit individuals from the population
def select_population(scores):
    temp = []
    i = 0
    while i < len(scores):
        temp.append(abs(target - scores[i]))
        i += 1
    least_fit = temp.index(max(temp))
    return least_fit





# Defining a function to crossover between two individual populations
def crossover(x, y):
    crossing = []
    crossover_point = random.randint(1, total_transactions - 2)
    fst_half = []
    lst_half = []
    tmp_fst_half = []
    tmp_lst_half = []

    i = 0
    while i <= crossover_point:
        fst_half.append(x[i])
        tmp_fst_half.append(y[i])
        i += 1

    i = crossover_point + 1
    while i < total_transactions:
        lst_half.append(x[i])
        tmp_lst_half.append(y[i])
        i += 1

    crossing.append(fst_half + tmp_lst_half)
    crossing.append(tmp_fst_half + lst_half)
    return crossing





# mutating ind in the population
def mutate_population(child1, child2):
    mutation_point = random.randint(0, total_transactions - 1)
    if child1[mutation_point] == 1:
        child1[mutation_point] = 0
    else:
        child1[mutation_point] = 1
    if child2[mutation_point] == 1:
        child2[mutation_point] = 0
    else:
        child2[mutation_point] = 1
    return child1, child2




# Initilizng an empty list for result
result = []

for i in range(50):

    # Generating a new population
    population = generate_population(total_transactions)

    # Calculating the fitness of each individual in the population
    scores = calculate_fitness(population)

    # Select the most fit individuals from the population
    least_fit = select_population(scores)

    # Removing the least fit individual from the population
    population.pop(least_fit)

    # Find the solution
    found_solution = False

    for j in range(len(population)):

        # Select two individuals at random to crossbreed
        x = random.choice(population)
        y = random.choice(population)

        # Generating two new children through crossover
        child = crossover(x, y)
        child1, child2 = child[0], child[1]

        # Mutate the children
        child1, child2 = mutate_population(child1, child2)

        # Check if the children have a sum equal to the target
        sum = 0
        for x, el in enumerate(child1):
            if el == 1:
                sum += amount[x]
        if sum == target:
            if child1 != ([0] * total_transactions):
                result = child1
                found_solution = True
                break
        sum = 0
        for x, el in enumerate(child2):
            if el == 1:
                sum += amount[x]
        if sum == target:
            if child2 != ([0] * total_transactions):
                result = child2
                found_solution = True
                break

    # If a solution is found, break out of the loop
    if found_solution == True:
        break

# If no solution is found, print -1
if len(result) == 0:
    print('-1')

# Otherwise, print the solution
else:
    for i in range(len(result)):
        print(result[i], end='')

```
