```python
def notation(n):
    if n=='Arad': return 'A'
    elif n=='Bucharest': return 'Z'
    elif n=='Craiova': return 'S'
    elif n=='Eforie': return 'T'
    elif n=='Fagaras': return 'O'
    elif n=='Dobreta': return 'V'
    elif n=='Hirsova': return 'N'
    elif n=='lasi': return 'Q'
    elif n=='Neamt': return 'F'
    elif n=='Oradea': return 'B'
    elif n=='Pitesti': return 'P'
    elif n=='RimnicuVilcea': return 'R'
    elif n=='Timisoara': return 'C'
    elif n=='Urziceni': return 'D'
    elif n=='Vaslui': return 'H'
    elif n=='Zerind': return'E'
    elif n=='Lugoj': return 'G'
    elif n=='Mehadia': return 'L'
    elif n=='Sibiu': return 'J'
    elif n=='Giurgiu': return 'K'
    
def ReverseNotation(n):
    if n=='A': return 'Arad'
    elif n=='Z': return 'Bucharest'
    elif n=='S': return 'Craiova'
    elif n=='T': return 'Eforie'
    elif n=='O': return 'Fagaras'
    elif n=='V': return 'Dobreta'
    elif n=='N': return 'Hirsova'
    elif n=='Q': return 'lasi'
    elif n=='F': return 'Neamt'
    elif n=='B': return 'Oradea'
    elif n=='P': return 'Pitesti'
    elif n=='R': return 'Rimnicu Vilcea'
    elif n=='C': return 'Timisoara'
    elif n=='D': return 'Urziceni'
    elif n=='H': return 'Vaslui'
    elif n=='E': return'Zerind'
    elif n=='G': return 'Lugoj'
    elif n=='L': return 'Mehadia'
    elif n=='J': return 'Sibiu'
    elif n=='K': return 'Giurgiu'
    
def aStar(start_node, stop_node):
    open_list=[]
    open_list.append(start_node)
    close_list=[]
    g={}
    parents={}
    g[start_node]=0
    parents[start_node]=start_node
    
    while len(open_list)!=0:
        n=None
        for v in open_list:
            if n==None or g[v]+h_dist[v] < g[n]+h_dist[n]:
                n=v
        if n==stop_node or nodes[n]==None:
            pass
        else:
            for (m, weight) in child(n):
                if m not in open_list and m not in close_list:
                    open_list.append(m)
                    parents[m] = n
                    g[m] = g[n]+weight

                else:
                    if g[m] > g[n]+weight:
                        g[m] = g[n]+weight
                        parents[m] = n
                        if m in close_list:
                            close_list.remove(m)
                            open_list.append(m)
        if n == None:
            print('NO PATH FOUND')
            return None

        if n==stop_node:
            path = []
            while parents[n]!=n:
                path.append(n)
                n = parents[n]
            path.append(start_node)
            path.reverse()
            return path,g

        open_list.remove(n)
        close_list.append(n)
    print('NO PATH FOUND')
    return None

def child(c):
    if c in nodes:
        return nodes[c]
    else:
        return None

input_file=open('Input file.txt', 'r')
lines=input_file.readlines()

nodes={}
h_dist={}

for i in range(len(lines)):
    if i!=(len(lines)-1):
        items = lines[i][:-1].split(' ')
        
    else:
        items = lines[i].split(' ')
        
    for i in range(len(items)):
        try:
            items[i]=int(items[i])
        except:
            items[i]=notation(items[i])
    
    name=[]
    val=[]
    for i in range(2,len(items)):
        if i%2==0:
            name.append(items[i])
        else:
            val.append(items[i])
            
    temp=[]
    for i in range(len(name)):
        temp2=[]
        temp2.append(name[i])
        temp2.append(val[i])
        temp.append(temp2)
    
    nodes.update({items[0]:temp})
    h_dist.update({items[0]:int(items[1])})


start_input=input('Start Node: ')
end_input=input('Destination: ')
start_node=notation(start_input)
end_node=notation(end_input)

outputs=aStar(start_node, end_node)

path_nodes=outputs[0]
path=''
for i in range(len(path_nodes)):
    if i!=(len(path_nodes)-1):
        path+=(ReverseNotation(path_nodes[i])+' -> ')
    else:
        path+=ReverseNotation(path_nodes[i])
print('Path: '+str(path))

path_distances=outputs[1]
temp=[]
for i in path_distances.values():
    temp.append(i)
distance=temp[-1]
print('Total distance: '+str(distance)+'km')

input_file.close()
```

```python
import math
import random

def alpha_beta_pruning(b, d, n_depth, n_position, leafs, alpha, beta, maximizing):
    global pruned_node
    if n_depth == d:
        return leafs[n_position]
 
    if maximizing == True:
        maxDamage = -math.inf
        for i in range(0, b):
            temp = alpha_beta_pruning(b, d, n_depth+1, (n_position*2)+i, leafs, alpha, beta, False)
            maxDamage = max(maxDamage, temp)
            alpha = max(alpha, maxDamage)
            if alpha >= beta:
                pruned_node+=1
                break
        return maxDamage
      
    else:
        maxDamage = math.inf
        for i in range(0, b):
            temp = alpha_beta_pruning(b, d, n_depth+1, (n_position*2)+i, leafs, alpha, beta, True)
            maxDamage = min(maxDamage, temp)
            beta = min(beta, maxDamage)
            if alpha >= beta:
                pruned_node+=1
                break
        return maxDamage

#======================================================================================

user_in1 = input('Enter your student id: ')
user_in2 = input('Minimum and Maximum value for the range of negative HP: ')
b = int(user_in1[2])
d = 2*int(user_in1[0])
hp = int(user_in1[:5:-1])
pruned_node = 0
leaf_count = b**d
l_range1, l_range2 = user_in2.split()
leafs=[]

for i in range(leaf_count):
    temp = random.randint(int(l_range1),int(l_range2))
    leafs.append(temp)

negative_hp = alpha_beta_pruning(b, d, 0, 0, leafs, -math.inf, math.inf, True)

#======================================================================================

out_leafs = ''
for i in range(len(leafs)):
    if i!=(len(leafs)-1):
        out_leafs += str(leafs[i])+', '
    else:
        out_leafs += str(leafs[i])
        
print('1. Depth and Branches ratio is '+str(d)+':'+str(b))
print('2. Terminal States(Leaf Nodes) are', out_leafs)
print('3. Left life(HP) of the defender after maximum damage caused by the attacker is', hp-negative_hp)
print('4. After Alpha-Beta Pruning Leaf Node Comparisons', leaf_count-pruned_node)
```
