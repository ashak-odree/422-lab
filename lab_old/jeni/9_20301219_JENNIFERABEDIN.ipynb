```python
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                      from random import randint
from math import ceil

def randomarray(minimum_point, maximum_point, length):
    array = []
    for i in range(length):
        array.append(randint(minimum_point, maximum_point))
    return array

def arrayshuffle(array):
    for i in range(randint(1,len(array))):
        first_index = randint(0,len(array)-1)
        second_index = randint(0,len(array)-1)
        array[first_index], array[second_index] = array[second_index], array[first_index]
    return array

def alphabetapruning(array, index = 0 , level = 0, level_state = "MAX", alpha = float('-inf'), beta = float('inf')):
    if level == 3:
        return array[index]
    if level_state == "MIN":
        left = alphabetapruning(array, index * 2, level + 1, "MAX", alpha, beta)
        beta = min(beta, left)
        if beta <= alpha:
            return left #purned
        right = alphabetapruning(array, index * 2 + 1, level + 1, "MAX", alpha, beta)
        minimum_value = min(left, right)
        return minimum_value
    else:
        left = alphabetapruning(array, index * 2, level + 1, "MIN", alpha, beta)
        alpha = max(alpha, left)
        if beta <= alpha:
            return left
        right = alphabetapruning(array, index * 2 + 1, level + 1, "MIN", alpha, beta)
        maximum_value = max(left, right)
        return maximum_value

std_id = '28381219' 
minimum_point = int(std_id[4]) 
points_to_win = int(std_id[1:3])
maximum_point = ceil(points_to_win * 1.5)
n_shuffles = int(std_id[3])
array = randomarray(minimum_point, maximum_point, 8)
achived_point = alphabetapruning(array)

if achived_point >= points_to_win:
    print("The winner is Optimus Prime")
else:
    print("The winner is Megatron")

List1 = []
count = 0
for i in range(n_shuffles):
    array = arrayshuffle(array)
    points = alphabetapruning(array)
    if points >= points_to_win:
        count += 1
    List1.append(points)
    
print("Generated 8 random points between the minimum and maximum point limits:",array)
print("total points to win:",points_to_win)
print("Achieved point by applying alpha-beta pruning =",achived_point)
print("After the shuffle:")
print("List of all points values from each shuffle:",List1)
print("The maximum value of all shuffles", max(List1))
print("Won",count,'times out of',n_shuffles,"number of shuffles")
```
