```python
fe = open("/content/Input file.txt","r")
f=fe.read().split('\n')
graph = {}
heuristics = {}
for line in f:
    line = line.strip()     
    line = line.split(' ')

    heuristics[line[0]] = int(line[1])
    graph[line[0]] = [[line[i] for i in range(2, len(line), 2)],[line[i] for i in range(3, len(line), 2)]]

def A_star(graph, h, start, end):
    queue = []
    reached = False
    path = ''
    distance = 0
    while not reached:
        for i in range(len(graph[start][0])):
            if len(queue) != 0:
                if queue[0][0] != graph[start][0][i]:
                    queue.append([start, graph[start][0][i], int(graph[start][1][i])])
            else:
                queue.append([start, graph[start][0][i], int(graph[start][1][i])])
        queue.sort(key=lambda x: x[2] + h[x[1]])
        start = queue[0][1]
        path += queue[0][0] + ' -> '
        distance += queue[0][2]
        if start == end:
            reached = True
            path += start
            print('Path: ' + path)
            print('Distance: ' + str(distance) + ' km')
            return path , distance
        queue.pop(0)

start = input('Start node: ')
goal = input('Destination: ')

print(A_star(graph, heuristics, start, goal))
```
