```python
def inp_graph(inp_file): #will read from the file that was given as 
                         #input
        graph={} #will store the graph here so taking an empty dict
        with open(inp_file, 'r') as file: #opening file
            for line in file: #will iterate each line in the file
                text = line.split() #splitting
                node = text[0]
                hurst = int(text[1])    
                edges = {}
                for i in range(2, len(text), 2):
                    edge = text[i]
                    d= int(text[i+1])
                    edges[edge] = d
                graph[node] = {"h":hurst, "e":edges} #in dict graph first will be node 
                                         # then huristic value then, the edges
                                         # of node
        return graph #will return the graph

def a_star(start, goal, graph):
    #setting the values at first
    open_set = {start}
    closed_set = set()
    
    cost_s = {start: 0} #cost to reach each n
    f_cost = {start: graph[start]['h']} #final estimated cost
    track = {} # keeping track of the cost nodes from parent nodes(pth)
    
     
    while open_set:
        #will get node with min f_cost
        c_node = min(open_set,key=lambda node: f_cost[node])

        # checking if reached to goal node
        if c_node == goal:
            path = [c_node]
            while c_node in track:
                c_node = track[c_node]
                path.append(c_node)
            path.reverse()
            return (f_cost[goal], path) #will return final cost and the path
        
        open_set.remove(c_node)
        closed_set.add(c_node) #will move the c_node to closed set
         # will check edges closed set
        for edge, d in graph[c_node]['e'].items():
            
            if edge in closed_set: #edge will be avoided
                continue

            # will Calculate temporary tenative cost_s
            u_cost_s = cost_s[c_node] + d
            
            if edge not in open_set: # adding edges if not aready added
                open_set.add(edge)

            elif u_cost_s >= cost_s[edge]: #cost update
                continue

            
            track[edge] = c_node #path update according to best
            cost_s[edge] = u_cost_s
            f_cost[edge] = cost_s[edge] + graph[edge]['h']

    
    return None #if no path


graph = inp_graph('C://Users/Asus/Downloads/input0.txt') #reading and printing graph
print(graph)
start = input("Start node: ") #taking user input
goal = input("Destination: ")
result = a_star(start, goal, graph)
if result is None:
    print( "NO PATH FOUND")
else:
    d, path = result
    print("Path:", " -> ".join(path))
    print("Total Distance:", d,"KM")
  



        
    


```
