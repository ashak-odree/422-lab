```python
import math
import random


def abPruning(node,dpth,alpha,beta,attkr,bullet):#defining the algorithm
    #checking if the given node is leaf node or max depth
    if dpth == 0 or node in terminal_nodes: ###
        return node
    if attkr: #getting thex max
        hp = -math.inf
        for i in getChild(node, bullet): #for all child node get max value
            hp = max(hp, abPruning(i, dpth - 1, alpha, beta, False, bullet))
            alpha = max(alpha, hp)
            
            if beta <= alpha: #when beta becomes less or equal prune
                increment_p() #keeping pruning count by calling
                break
                
        return hp
    else:
        hp = math.inf #for the min value
        for i in getChild(node, bullet): #for each child node gen min value
            hp = min(hp, abPruning(i, dpth - 1, alpha, beta, True, bullet))
            
            beta = min(beta, hp)
            if beta <= alpha:#when beta becomes less or equal prue
                increment_p() #keeping pruning count by calling
                break
                
        return hp
p=0
def increment_p():#prune counter function
    global p
    p += 1
    
def getChild(node, bullet): #get child from given node input
    child = []
    for i in bullet: # Subtracting each bullet from node to get the child node
        child.append(node - i)
    return child

#getting all inputs
student_id = input("Enter your student id: ")
min_hp, max_hp = map(int, input("Minimum and Maximum value for the range of negative HP: ").split())
chance = int(student_id[0])#first chance
dpth =chance*2 #both attacker and defender will get chance
defender_hp = int(student_id[::-1][:2]) #reverse and last to digit
num_bullet = int(student_id[2])
terminal_nodes = random.sample(range(min_hp, max_hp+1), pow(dpth,num_bullet))

bullet=[] #getting bullets from input
for i in range(1,num_bullet+1):
    bullet.append(i)
    
totalnodes = pow(dpth,num_bullet) #total nodes for the depth



branches = num_bullet

depthBranches_ratio = str(dpth) + ":" + str(branches) #ratio for the output


max_damage = abPruning(defender_hp, dpth, max_hp, min_hp, True, bullet) #getting the damage that 
#attacker deals

remainig_life = defender_hp - max_damage

#printing all output

print("Depth and Branches ratio is", depthBranches_ratio)
print("Terminal States(Leaf Nodes) are"," , ".join(map(str,terminal_nodes)))
print("Left life(HP) of the defender after maximum damage caused by the attacker is", int(remainig_life))
print("After Alpha-Beta Pruning Leaf Node Comparisons",totalnodes-int(p))


```
