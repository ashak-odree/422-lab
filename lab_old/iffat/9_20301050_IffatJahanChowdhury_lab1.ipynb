```python
from collections import deque

inp_file = open('C:\\Users\\HP\\OneDrive\\Desktop\\422_lab1_input.txt','r')
inp_lines = inp_file.readlines()

dict1 = {}
dict2 = {}

for i in inp_lines:
    l1 = i[:-1].split(' ')
    l2 = l1[2:]
    dict1[l1[0]] = []
    for j in range(len(l2)-1):
        if j%2==0:
            dict1[l1[0]].append((l2[j],int(l2[j+1])))
    dict2[l1[0]]=int(l1[1])
    
    
    
class Graph:
    
    def __init__(self, adj_list,heu_val):
        self.adj_list = adj_list
        self.heu_val = heu_val
        
    def get_neighbors(self, k):
        return self.adj_list[k]
    
    def heu(self, n):
        return self.heu_val[n]
    
    def a_star_algorithm(self, starting_node, ending_node):
        open_list = set([starting_node])
        closed_list = set([])
        r= {}
        r[starting_node] = 0
        parents = {}
        parents[starting_node] = starting_node
        
        
        while len(open_list) > 0:
            n = None
            
            for v in open_list:
                if n == None or r[v] + self.heu(v) < r[n] + self.heu(n):
                    n = v;
                    
            if n == None:
                print('NO PATH FOUND')
                return None
            
            if n == ending_node:
                reconst_path = []
                
                while parents[n] != n:
                    reconst_path.append(n)
                    n = parents[n]
                    
                reconst_path.append(start_node)
                reconst_path.reverse()
                distance = 0
                
                for x in range(len(reconst_path)-1):
                    for (y,z) in self.adj_list[reconst_path[x]]:
                        if y==reconst_path[x+1]:
                            distance+=z
                            
                return (f"Path: {' ->' .join(reconst_path)}\nTotal distance: {distance} km")
            
            
            for (m, weight) in self.get_neighbors(n):
                if m not in open_list and m not in closed_list:
                    open_list.add(m)
                    parents[m] = n
                    r[m] = r[n] + weight
                    
                else:
                    if r[m] > r[n] + weight:
                        r[m] = r[n] + weight
                        parents[m] = n
                        if m in closed_list:
                            closed_list.remove(m)
                            open_list.add(m)
                            
                            
            open_list.remove(n)
            closed_list.add(n)
            
        print('NO PATH FOUND')
        return None

    
    
starting_node = input('Start node: ')
ending_node = input('Destination: ')
graph1 = Graph(dict1,dict2)
print(graph1.a_star_algorithm(starting_node, ending_node))
```
