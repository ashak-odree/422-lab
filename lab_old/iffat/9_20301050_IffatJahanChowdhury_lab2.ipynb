```python
import math
import random


id = input('Student ID: ')
print()

if len(id) != 8:
    raise ValueError('Student ID must be of 8 digits')

if '0' in id:
    id = id.replace('0', '8')

SHUFFLE = int(id[3])
LEN, MIN = 8, int(id[4])
TARGET = int(id[-1] + id[-2])
MAX = int(TARGET * 1.5)
INF = math.inf
leaves = random.sample(range(MIN, MAX), LEN)  


def pruning(n, d, a, b, f, m):    #n-->node, d-->depth, a-->alpha, b-->beta, f-->leaves(list), m-->maximize(bool)
    """
    It is a recursive function that performs alpha-beta
    pruning on the leaf-nodes and returns the optimal
    value. 
    """

    if d == 3:
        return f[n]

    if m:  # Optimus Prime's Turn
        point = -INF
        for i in range(2):
            value = pruning(n * 2 + i,  
                            d + 1,
                            a, b,
                            f, False)
            point = max(point, value)  #initially point=-inf, -inf aar value er moddhe je ta boro oi ta point e 
            a = max(a, point)  
            
            if a >= b:
                break
        
        return point
        
    else:   # Megatron's Turn
        point = INF
        for i in range(2):
            value = pruning(n * 2 + i,
                            d + 1,
                            a, b,
                            f, True)
            point = min(point, value)
            b = min(b, point)
            
            if a >= b:
                break
            
        return point


if __name__ == '__main__':
    record = 0   #counter of op's winning
    scores = []
    
    for i in range(SHUFFLE):
        final = pruning(0, 0,         #a-b pruning er por top node e jei value ta 
                        -INF, INF,
                        leaves, True)
        
        if final >= TARGET:
            winner = 'Optimus Prime'
            record += 1
        else:
            winner = 'Megatron'
        
        print('Generated 8 random points between the minimum & maximum limits:')
        print(str(leaves)[1:-1])
        print(f'Total points to win: {TARGET}')
        print(f'Achieved point by applying Alpha-Beta Pruning: {final}')
        print(f'Winner is {winner}')
        print()
        
        scores.append(final) 
        random.shuffle(leaves)
    
    print('After the shuffle,')
    print('List of all point values from each shuffle:')
    print(str(scores)[1:-1])
    print(f'The maximum value of all shuffles: {max(scores)}')
    print(f'Optimus Prime won {record} times out of {SHUFFLE} number of shuffles')
```
