```python
import random as rd
import numpy as np
from queue import Queue


''''
5 240
Bradman 120
Tendulkar 90
Sangakkara 70
Kallis 65
Lara 80

Sample Output 2
['Bradman', 'Tendulkar', 'Sangakkara', 'Kallis', 'Lara']
01101

''''

COUNT, TARGET = map(int, input().split())   #integer e cast..map
GEN = Queue()  
INFO = []
LENGTH = 10


for i in range(COUNT):        #COUNT, TARGET = map(int, input().split())  
    name, run = input().split()
    INFO.append((name, int(run)))  #tuple


print(INFO)   #   [('Ralph', 80), ('John', 70), ('Tom', 40), ('Young', 50)]
print([INFO[i][0] for i in range(len(INFO))])   #player's name print



def populate(length):
    
    for x in range(length):
        gene = np.random.choice([0, 1], size=len(INFO))  #producing genes    info--> no. of batsman
        GEN.put(str(gene)[1:-1].replace(" ", ""))



def fit(gene):
    
    fitness = 0
    for x in range(len(gene)):
        if gene[x] == '1':
            fitness += INFO[x][1]
            
    return fitness == TARGET #bool return



def cross(gene_1, gene_2):
    
    x = rd.randint(0, len(gene_1)-1)  #cross over er jonno
    gene_3 = gene_1[:x] + gene_2[x:]
    gene_4 = gene_2[:x] + gene_1[x:]
    
    return gene_3, gene_4



def mutate(gene):
    
    gene = [int(i) for i in gene]
    m = rd.randint(0, len(gene) // 2 - 1)
    n = rd.randint(len(gene) // 2, len(gene) - 1)
    gene[m], gene[n] = gene[n], gene[m]  
    
    return "".join(str(i) for i in gene)



if __name__ == "__main__":
    populate(LENGTH)
    found = False
    for i in range(10**5):
        if not found:
            gene_1, gene_2 = GEN.get(), GEN.get()
            if fit(gene_1):
                print(gene_1)
                found = True
                break
            elif fit(gene_2):
                print(gene_2)
                found = True
                break
            else:
                gene_3, gene_4 = cross(gene_1, gene_2)
                GEN.put(mutate(gene_3))
                GEN.put(mutate(gene_4))


    print(-1) if not found else exit()
```
